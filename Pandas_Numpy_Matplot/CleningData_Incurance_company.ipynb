```python
file2 = pd.read_csv('Data/file2.csv')
file2['Number of Open Complaints'] = file2['Number of Open Complaints'].str.get(2)
file2
```

```python
#change names of columns
file3 = pd.read_csv('Data/file3.csv')
file3
file3.rename(columns = {'State' : 'ST', 'Gender' : 'GENDER'}, inplace = True) 
file3
```

```python
#Open and merge the files with correct columns names
def load_original_data():
    file1 
    file2 
    file3 
    return pd.concat([file1,file2,file3], axis=0)

df = load_original_data()
df
```

```python
#Make the column names lower case
def lower_case_column_names(df):
    df.columns=[i.lower() for i in df.columns]
    return df

df=lower_case_column_names(df)
df
```

```python
#removing duplicates
df.drop_duplicates()
df
```

```python
#deleting NaN values rows
#dont work - df.dropna(axis=0, how="any", thresh=None, subset=None, inplace=False)
df = df[df['customer lifetime value'].isna()==False]
df
```

```python
#check how much colums with NAN values
nulls_df = pd.DataFrame(round(df.isna().sum()/len(df),4)*100)
nulls_df = nulls_df.reset_index()
nulls_df.columns = ['header_name', 'percent_nulls']
nulls_df
```

```python
#deleting % from CLV column

#df = df[df['customer lifetime value']].replace(to_replace ='%', value = '', regex = True)
df['customer lifetime value'] = df['customer lifetime value'].replace(r'%', '', regex=True)
df
```

```python
#turn types to numeric dont need
df['customer lifetime value'] =  pd.to_numeric(df['customer lifetime value'], errors='coerce')
#to early for this df['number of open complaints'] =  pd.to_numeric(df['number of open complaints'], errors='coerce')
df
```

```python
#round for 2 colums
df['customer lifetime value'] = np.round(df['customer lifetime value'])
df['total claim amount'] = np.round(df['total claim amount'])
df
```

```python
#turn float to integer
df['customer lifetime value'] = df['customer lifetime value'].astype(int)
df['income'] = df['income'].astype(int)
df['monthly premium auto'] = df['monthly premium auto'].astype(int)
df['total claim amount'] = df['total claim amount'].astype(int)
df['number of open complaints'] = df['number of open complaints'].astype(int)
df
```

```python
#deleting Cust column
df = df.drop(columns = ['customer'],axis = 1)
df
```
